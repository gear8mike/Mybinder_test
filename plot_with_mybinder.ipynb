```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-3, 3, 0.01)
y = 2*np.sin(3*x) +1
y1 = 0*x + 2
plt.plot(x,y, c='r')
plt.plot(x,y1, color = 'b')
plt.fill_between(x,y, y1, color ='g', alpha=0.3,)
print('examples of plots with matplotlib')
```
